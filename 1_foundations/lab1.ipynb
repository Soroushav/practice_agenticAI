```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)

groq_api_key = os.getenv("GROQ_API_KEY")
gemini_api_key = os.getenv("GEMINI_API_KEY")

openai = OpenAI()
groq = OpenAI(api_key=groq_api_key, base_url="https://api.groq.com/openai/v1")
gemini = OpenAI(api_key=gemini_api_key, base_url="https://generativelanguage.googleapis.com/v1beta")
```

```python
messages = [{"role": "user", "content": "Hello, how are you?"}]

response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
    temperature=0.5,
    max_tokens=100,
)

output = response.choices[0].message.content
from IPython.display import Markdown, display

display(Markdown(output))


```

```python
content = """
    Goal: Create ONE challenging, high-discrimination question to compare AI models. The question must require deep reasoning, not trivia, and should be answerable without browsing.

Domain: [pick one:logic, planning, science, law-ish reasoning, creative constraints, etc.]
Difficulty target: hard for general LLMs, solvable by top models with careful thinking.
Constraints:
	1.	The question must be self-contained.
	2.	It should have a clear, checkable “ground truth” or at least an objectively gradable standard.
	3.	It should include at least 2 traps that catch shallow pattern-matching (e.g., misleading wording, hidden constraints, edge cases).
	4.	It must be possible to score answers on reasoning quality, not just final output.
"""
messages = [{"role": "user", "content": content}]

response = openai.chat.completions.create(
    model="gpt-5.1-2025-11-13",
    messages=messages
)

question = response.choices[0].message.content

display(Markdown(question))

```

```python
model_names = []
answers = []

messages = [{"role": "user", "content": question}]
```

```python
## Lets start with openAI

model_name = "gpt-4o-mini"

response = openai.chat.completions.create(
    model=model_name,
    messages=messages,
)

answer = response.choices[0].message.content

model_names.append(model_name)
answers.append(answer)

display(Markdown(answer))
```

```python
# now lets try groq

model_name = "llama-3.3-70b-versatile"

response = groq.chat.completions.create(
    model=model_name,
    messages=messages,
)

answer = response.choices[0].message.content

model_names.append(model_name)
answers.append(answer)

display(Markdown(answer))
```

```python
# now lets try gemini

model_name = "gemini-2.5-flash"

response = gemini.chat.completions.create(
    model=model_name,
    messages=messages
)

answer = response.choices[0].message.content

model_names.append(model_name)
answers.append(answer)

display(Markdown(answer))
```

```python
together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
display(Markdown(together))
```

```python
judge = f""" You are judging a competition between {len(model_names)} competitors.
    Each model has been given this question:
    {question}
    Your tasks

1. Extract each model’s claimed result

For each answer:
	•	List the experiments it claims can be in slot 5.
	•	List any experiments it claims cannot be in slot 5 (if mentioned).

2. Validate schedules

The benchmark requires that for each claimed possible experiment in slot 5, the model must give at least one explicit full schedule of 8 experiments showing it.

For every schedule the model provides:
Check it against all constraints:
Precedence
	•	A before D but not adjacent
	•	B after E but not adjacent
	•	C between F and G (either order)
	•	H either first OR immediately after G, but not both

Adjacency / separation
	•	No adjacency for (A,B), (C,D), (E,F)
	•	D and E exactly 3 apart
	•	F and G not in positions 1 or 8

Extra structural traps
	•	Trap 1: “no identical before/after pattern”
	•	Identify all ordered before/after pairs implied by rules.
	•	Verify that no two such pairs share identical (pos(X), pos(Y)).
	•	Trap 2: Exactly one of (i) or (ii) is true.
	•	Evaluate (i) and (ii) correctly based on whether H is first.
	•	Confirm XOR: one true, one false.

Mark each schedule as: VALID or INVALID, and give the first violated rule if invalid.

3. Decide correctness of each model’s final set

For each model:
	•	If it includes any impossible experiment for slot 5 → that’s a major error.
	•	If it misses a possible experiment → also a major error.
	•	If it gives correct set but invalid supporting schedules → set may be right, but reasoning/support is weak.
    
    Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
    Respond with JSON, and only JSON, with the following format:
    {{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}
    Here are the responses from each competitor:
    {together}

    Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks.
"""

messages = [{"role": "user", "content": judge}]
response = openai.chat.completions.create(
    model="gpt-5.1-2025-11-13",
    messages=messages,
)

result = response.choices[0].message.content

display(Markdown(result))


```

```python
import json


results_dict = json.loads(result)
for i, model_name in enumerate(model_names):

    print(f"Model {i+1}: {model_name}")
    print(f"Rank: {results_dict['results'][i]}\n")


```
